```python
import pandas as pd
import random

sabores = [
    "Abacaxi", "Açaí", "Amêndoa", "Baunilha", "Blue Ice", "Canela", "Caramelo", "Caramelo Salgado",
    "Cereja", "Chocolate", "Chocolate Branco", "Coco", "Cookies and Cream", "Damasco", "Doce de leite",
    "Framboesa", "Fresa", "Frutas vermelhas", "Gengibre", "Groselha", "Hortelã", "Iogurte", "Jabuticaba",
    "Kiwi", "Laranja", "Limão", "Limão Siciliano", "Maçã", "Manteiga de amendoim", "Maracujá", "Melancia",
    "Melão", "Menta", "Morango", "Morango com Banana", "Nozes", "Nutella", "Pera", "Pêssego", "Pistache",
    "Pitanga", "Romã", "Tangerina", "Tutti-frutti", "Uva", "Violeta", "Waffle", "Xarope de bordo", "Yuzu",
    "Zimbro"
]

valores = [round(random.uniform(5, 15), 2) for _ in range(len(sabores))]
estoque = [round(random.uniform(50, 200), 2) for _ in range(len(sabores))]

dados = {'Sabores': sabores, 'Valores': valores, 'Quantidade Estoque (Kg)': estoque}
df = pd.DataFrame(dados)

print(df)
```
